```python
# ─────────────────────────────────────────────────────────────────────────────
# Reemplazo de valores "no respuesta" por NaN en las variables de interés
# dentro de df_Mod_Vic (tabla TMod_Vic)
# ─────────────────────────────────────────────────────────────────────────────

# Mapeo: variable → valores que deben convertirse a NaN
no_respuesta = {
    # Hora del delito: 9 = No sabe/no responde; 'b' = blanco (no aplica)
    'BP1_4':  [9, 'b'],

    # Lugar del delito: 9 = No sabe/no responde
    'BP1_5':  [9],

    # Nº de delincuentes: 9 = No sabe/no responde; 'b' = blanco
    'BP1_9':  [9, 'b'],

    # ¿Llevaban arma?: 9 = No sabe/no responde; 'b' = blanco
    'BP1_15': [9, 'b'],

    # ¿Lesión con el arma?: 9 = No sabe/no responde; 'b' = blanco
    'BP1_17': [9, 'b'],

    # ¿Otro tipo de violencia física?: 9 = No sabe/no responde; 'b' = blanco
    'BP1_18': [9, 'b'],

    # Daño principal sufrido: 9 = No sabe/no responde
    'BP1_33': [9],
}

# BP1_16_* son variables flag (0/1) que indican el tipo de arma.
# Solo tienen sentido cuando BP1_15 == 1 (sí llevaban arma).
# Cuando BP1_15 != 1 el campo queda en blanco ('b') → ya es NaN en muchos casos,
# pero lo forzamos por si quedó como string.
bp1_16_cols = [c for c in df_Mod_Vic.columns if c.startswith('BP1_16_')]

for col, valores in no_respuesta.items():
    if col in df_Mod_Vic.columns:
        df_Mod_Vic[col] = df_Mod_Vic[col].replace(valores, np.nan)

for col in bp1_16_cols:
    df_Mod_Vic[col] = df_Mod_Vic[col].replace('b', np.nan)

# ─────────────────────────────────────────────────────────────────────────────
# Verificación rápida: conteo de NaN por variable
# ─────────────────────────────────────────────────────────────────────────────
vars_check = list(no_respuesta.keys()) + bp1_16_cols
print(df_Mod_Vic[vars_check].isna().sum().to_string())
```

```python
# ------------------------------------------------------------
# Funciones auxiliares
# ------------------------------------------------------------

def limpiar_a_numerico(s):
    """
    Convierte valores tipo 'b', blancos, True/False y strings numéricos
    a códigos numéricos utilizables.
    """
    s = s.replace({
        True: 1,
        False: 0,
        "True": 1,
        "False": 0,
        "true": 1,
        "false": 0,
        "TRUE": 1,
        "FALSE": 0,
        "b": np.nan,
        "B": np.nan,
        "": np.nan,
        " ": np.nan,
        "nan": np.nan,
        "NaN": np.nan,
        "None": np.nan,
        None: np.nan
    })
    
    return pd.to_numeric(s, errors="coerce")


def recodificar_directo(df, cols, mapa):
    """
    Recodifica directamente en el dataframe original.
    No crea columnas _bin.
    """
    cols_existentes = [c for c in cols if c in df.columns]
    
    for c in cols_existentes:
        s = limpiar_a_numerico(df[c])
        df[c] = s.map(mapa).astype(float)
    
    return df


def limpiar_conteos(df, cols):
    """
    Para variables de cantidad, por ejemplo AP7_4_*.
    Convierte 'b' y otros valores no numéricos a NaN.
    """
    cols_existentes = [c for c in cols if c in df.columns]
    
    for c in cols_existentes:
        df[c] = limpiar_a_numerico(df[c]).astype(float)
    
    return df


def missing_percent(df, cols):
    cols = [c for c in cols if c in df.columns]
    return (
        df[cols]
        .isna()
        .mean()
        .mul(100)
        .round(2)
        .sort_values(ascending=False)
    )


def plot_missing_bar(missing_series, title, figsize=(12, 6)):
    if missing_series.empty:
        print(f"No hay variables para graficar en: {title}")
        return
    
    data = missing_series.sort_values(ascending=True)
    
    cmap = plt.cm.RdYlGn_r
    norm = plt.Normalize(vmin=0, vmax=100)
    colors = cmap(norm(data.values))
    
    plt.figure(figsize=figsize)
    plt.barh(data.index, data.values, color=colors)
    plt.xlabel("% de valores faltantes")
    plt.title(title)
    plt.xlim(0, 100)
    
    for i, v in enumerate(data.values):
        plt.text(v + 1, i, f"{v:.1f}%", va="center", fontsize=9)
    
    plt.tight_layout()
    plt.show()


def plot_missing_heatmap_percent(missing_series, title, figsize=(14, 2.8)):
    if missing_series.empty:
        print(f"No hay variables para graficar en: {title}")
        return
    
    df_plot = missing_series.to_frame().T
    df_plot.index = ["% missing"]
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        df_plot,
        annot=True,
        fmt=".1f",
        cmap="RdYlGn_r",
        vmin=0,
        vmax=100,
        cbar_kws={"label": "% missing"}
    )
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
resultados = {}

# Mapas de recodificación directa
# Resultado final:
# 1 = presencia del fenómeno de interés
# 0 = ausencia del fenómeno de interés
# NaN = no aplica / no sabe / no responde / blanco

mapa_seguridad_geografica = {
    1: 0,        # Seguro
    2: 1,        # Inseguro
    9: np.nan    # No sabe / no responde
}

mapa_seguridad_espacios = {
    1: 0,        # Seguro
    2: 1,        # Inseguro
    3: np.nan,   # No aplica
    9: np.nan    # No sabe / no responde
}

mapa_caminar_noche = {
    1: 0,        # Muy seguro
    2: 0,        # Seguro
    3: 1,        # Inseguro
    4: 1,        # Muy inseguro
    5: np.nan,   # No aplica
    9: np.nan    # No sabe / no responde
}

mapa_si_no_noaplica = {
    1: 1,        # Sí
    2: 0,        # No
    3: np.nan,   # No aplica
    9: np.nan    # No sabe / no responde
}

mapa_si_no = {
    1: 1,        # Sí
    2: 0,        # No
    9: np.nan    # No sabe / no responde
}


for year, dfs in bases.items():
    
    per1 = dfs["per1"]
    per2 = dfs["per2"]
    
    # Por si ya habías corrido el bloque anterior y quedaron columnas _bin
    per1.drop(columns=[c for c in per1.columns if c.endswith("_bin")], inplace=True, errors="ignore")
    per2.drop(columns=[c for c in per2.columns if c.endswith("_bin")], inplace=True, errors="ignore")
    
    # ------------------------------------------------------------
    # Variables seleccionadas de percepción
    # ------------------------------------------------------------
    
    cols_ap43 = [
        "AP4_3_1",
        "AP4_3_2"
    ]
    
    cols_ap44 = [
        "AP4_4_01",
        "AP4_4_03",
        "AP4_4_05",
        "AP4_4_06",
        "AP4_4_07",
        "AP4_4_08",
        "AP4_4_09",
        "AP4_4_12"
    ]
    
    cols_ap44a = [
        "AP4_4_A"
    ]
    
    cols_ap410 = [
        "AP4_10_01",
        "AP4_10_02",
        "AP4_10_03",
        "AP4_10_04",
        "AP4_10_05",
        "AP4_10_06",
        "AP4_10_08",
        "AP4_10_09",
        "AP4_10_10",
        "AP4_10_11",
        "AP4_10_12",
        "AP4_10_13",
        "AP4_10_14",
        "AP4_10_15"
    ]
    
    # ------------------------------------------------------------
    # Variables AP7: victimización personal
    # ------------------------------------------------------------
    
    cols_ap73 = [c for c in per2.columns if c.startswith("AP7_3")]
    cols_ap74 = [c for c in per2.columns if c.startswith("AP7_4")]
    
    # ------------------------------------------------------------
    # Limpieza directa sobre matrices originales
    # ------------------------------------------------------------
    
    per1 = recodificar_directo(per1, cols_ap43, mapa_seguridad_geografica)
    per1 = recodificar_directo(per1, cols_ap44, mapa_seguridad_espacios)
    per1 = recodificar_directo(per1, cols_ap44a, mapa_caminar_noche)
    per1 = recodificar_directo(per1, cols_ap410, mapa_si_no_noaplica)
    
    per2 = recodificar_directo(per2, cols_ap73, mapa_si_no)
    per2 = limpiar_conteos(per2, cols_ap74)
    
    resultados[year] = {
        "per1": per1,
        "per2": per2,
        "cols_ap43": cols_ap43,
        "cols_ap44": cols_ap44,
        "cols_ap44a": cols_ap44a,
        "cols_ap410": cols_ap410,
        "cols_ap73": cols_ap73,
        "cols_ap74": cols_ap74
    }
```

```python
# ============================================================
# Matrices de correlación entre variables BP y AP
# ============================================================

# Etiquetas BP
labels_bp = {
    "BP1_4": "Hora",
    "BP1_9": "Nº delincuentes",
    "BP1_15": "Portó arma",
    "BP1_17": "Lesión c/arma",
    "BP1_18": "Otra violencia",
    "BP1_16_1": "Arma fuego",
    "BP1_16_2": "Arma blanca",
    "BP1_16_3": "Obj. contundente",
    "BP1_16_4": "Otra arma",
    "BP1_16_9": "Arma NS/NR"
}

mapa_lugar = {
    1: "Lugar: calle",
    2: "Lugar: casa",
    3: "Lugar: trabajo",
    4: "Lugar: negocio",
    5: "Lugar: público",
    6: "Lugar: transporte público",
    7: "Lugar: carretera",
    8: "Lugar: otro"
}

mapa_danio = {
    1: "Daño: económico",
    2: "Daño: físico",
    3: "Daño: emocional/psicológico",
    4: "Daño: laboral",
    5: "Daño: ninguno"
}


def preparar_bp_para_corr(mod):
    """
    Limpia BP directamente para matrices de correlación.
    - Binarias: 1 = presencia / sí, 0 = ausencia / no.
    - Ordinales: se conservan numéricas.
    - Nominales: BP1_5 y BP1_33 se convierten después a dummies.
    """
    df = mod.copy()
    
    # Variables BP principales
    bp_ordinales = ["BP1_4", "BP1_9"]
    bp_binarias = ["BP1_15", "BP1_17", "BP1_18"]
    bp_armas = ["BP1_16_1", "BP1_16_2", "BP1_16_3", "BP1_16_4", "BP1_16_9"]
    bp_nominales = ["BP1_5", "BP1_33"]
    
    # Conversión general a numérico
    for c in bp_ordinales + bp_binarias + bp_armas + bp_nominales:
        if c in df.columns:
            df[c] = (
                df[c]
                .replace({
                    True: 1, False: 0,
                    "True": 1, "False": 0,
                    "true": 1, "false": 0,
                    "b": np.nan, "B": np.nan,
                    "": np.nan, " ": np.nan,
                    "nan": np.nan, "NaN": np.nan,
                    "None": np.nan,
                    None: np.nan
                })
            )
            df[c] = pd.to_numeric(df[c], errors="coerce")
    
    # Ordinales: 9 = no sabe/no responde
    for c in bp_ordinales:
        if c in df.columns:
            df[c] = df[c].replace({9: np.nan})
    
    # Binarias sí/no: 1 = sí, 2 = no, 9 = NS/NR
    for c in bp_binarias:
        if c in df.columns:
            df[c] = df[c].replace({9: np.nan}).map({1: 1, 2: 0})
    
    # Flags de tipo de arma: 0/1; b ya quedó como NaN
    # Si BP1_15 == 2, es decir, no portó arma, las armas específicas pasan a 0.
    if "BP1_15" in df.columns:
        for c in bp_armas:
            if c in df.columns:
                df.loc[df["BP1_15"] == 0, c] = 0
                df[c] = df[c].replace({9: np.nan})
    
    # Nominales: 9 = NS/NR
    for c in bp_nominales:
        if c in df.columns:
            df[c] = df[c].replace({9: np.nan})
    
    return df


def agregar_dummies_nominales(df):
    """
    Convierte BP1_5 y BP1_33 a dummies.
    Esto evita correlacionar códigos nominales como si fueran escala.
    """
    df = df.copy()
    dummy_cols = []
    dummy_labels = {}
    
    if "BP1_5" in df.columns:
        lugar_cat = df["BP1_5"].map(mapa_lugar)
        d_lugar = pd.get_dummies(lugar_cat, dtype=float)
        d_lugar.loc[lugar_cat.isna(), :] = np.nan
        
        df = pd.concat([df, d_lugar], axis=1)
        dummy_cols += list(d_lugar.columns)
        dummy_labels.update({c: c for c in d_lugar.columns})
    
    if "BP1_33" in df.columns:
        danio_cat = df["BP1_33"].map(mapa_danio)
        d_danio = pd.get_dummies(danio_cat, dtype=float)
        d_danio.loc[danio_cat.isna(), :] = np.nan
        
        df = pd.concat([df, d_danio], axis=1)
        dummy_cols += list(d_danio.columns)
        dummy_labels.update({c: c for c in d_danio.columns})
    
    return df, dummy_cols, dummy_labels


def detectar_llaves(per1, mod):
    """
    Detecta llaves para unir TPer_Vic1 con TMod_Vic.
    Preferimos ID_PER si está disponible.
    """
    if "ID_PER" in per1.columns and "ID_PER" in mod.columns:
        return ["ID_PER"]
    
    posibles = ["UPM", "VIV_SEL", "HOGAR", "R_SEL"]
    return [c for c in posibles if c in per1.columns and c in mod.columns]


def preparar_base_corr_bp_ap(year, resultados, bases):
    """
    Une AP con BP.
    La unidad resultante es delito/evento de TMod_Vic.
    Las AP se repiten si una persona tiene más de un delito registrado.
    """
    per1 = resultados[year]["per1"].copy()
    mod = bases[year]["mod"].copy()
    
    mod = preparar_bp_para_corr(mod)
    mod, bp_dummies, dummy_labels = agregar_dummies_nominales(mod)
    
    # Variables AP
    ap_percepcion = (
        resultados[year].get("cols_ap43", []) +
        resultados[year].get("cols_ap44", []) +
        resultados[year].get("cols_ap44a", [])
    )
    
    ap_miedo = resultados[year].get("cols_ap410", [])
    
    ap_percepcion = [c for c in ap_percepcion if c in per1.columns]
    ap_miedo = [c for c in ap_miedo if c in per1.columns]
    
    # Variables BP correlacionables sin nominales originales
    bp_correlacionables = [
        "BP1_4",
        "BP1_9",
        "BP1_15",
        "BP1_17",
        "BP1_18",
        "BP1_16_1",
        "BP1_16_2",
        "BP1_16_3",
        "BP1_16_4"
    ]
    
    # Si querés incluir también la dummy de NS/NR de arma, descomentá esta línea:
    # bp_correlacionables.append("BP1_16_9")
    
    bp_correlacionables = [c for c in bp_correlacionables if c in mod.columns]
    
    llaves = detectar_llaves(per1, mod)
    
    per1_reducida = per1[llaves + ap_percepcion + ap_miedo].drop_duplicates(subset=llaves)
    mod_reducida = mod[llaves + bp_correlacionables + bp_dummies].copy()
    
    base_corr = mod_reducida.merge(
        per1_reducida,
        on=llaves,
        how="inner"
    )
    
    all_labels = {}
    all_labels.update(labels)
    all_labels.update(labels_bp)
    all_labels.update(dummy_labels)
    
    return base_corr, bp_correlacionables, bp_dummies, ap_percepcion, ap_miedo, all_labels


def corr_rectangular(df, filas, columnas, method="spearman", min_periods=30):
    """
    Matriz rectangular: filas BP x columnas AP.
    method='spearman' es recomendable para ordinales/binarias.
    """
    filas = [c for c in filas if c in df.columns]
    columnas = [c for c in columnas if c in df.columns]
    
    corr = df[filas + columnas].corr(
        method=method,
        min_periods=min_periods
    )
    
    return corr.loc[filas, columnas]


def plot_corr(corr, title, labels_dict=None, figsize=None, annot=True):
    """
    Grafica heatmap de correlaciones.
    """
    if corr.empty:
        print(f"Matriz vacía: {title}")
        return
    
    plot_df = corr.copy()
    
    if labels_dict is not None:
        plot_df = plot_df.rename(
            index=lambda x: labels_dict.get(x, x),
            columns=lambda x: labels_dict.get(x, x)
        )
    
    if figsize is None:
        figsize = (
            max(10, 0.55 * plot_df.shape[1]),
            max(5, 0.45 * plot_df.shape[0])
        )
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        plot_df,
        annot=annot,
        fmt=".2f",
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        center=0,
        linewidths=0.5,
        linecolor="white",
        cbar_kws={"label": "correlación"}
    )
    
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()


def top_correlaciones(corr, n=15):
    """
    Devuelve las correlaciones más altas en valor absoluto.
    Sirve para resumir matrices grandes.
    """
    return (
        corr
        .stack()
        .rename("correlacion")
        .reset_index()
        .rename(columns={"level_0": "BP", "level_1": "AP"})
        .assign(abs_corr=lambda x: x["correlacion"].abs())
        .sort_values("abs_corr", ascending=False)
        .head(n)
    )
```

```python
# ============================================================
# Generar matrices por año
# ============================================================

for year in ["2024", "2025"]:
    
    base_corr, bp_corr, bp_dummies, ap_percepcion, ap_miedo, all_labels = preparar_base_corr_bp_ap(
        year,
        resultados,
        bases
    )
    
    print(f"\n==================== ENVIPE {year} ====================")
    print(f"Base para correlación: {base_corr.shape[0]} filas x {base_corr.shape[1]} columnas")
    
    # --------------------------------------------------------
    # 1) BP principales x AP percepción de inseguridad
    # --------------------------------------------------------
    corr_bp_ap_percepcion = corr_rectangular(
        base_corr,
        filas=bp_corr,
        columnas=ap_percepcion,
        method="spearman"
    )
    
    plot_corr(
        corr_bp_ap_percepcion,
        title=f"{year} - BP principales x AP percepción de inseguridad",
        labels_dict=all_labels,
        figsize=(12, 6)
    )
    
    # --------------------------------------------------------
    # 2) BP principales x AP miedo conductual
    # --------------------------------------------------------
    corr_bp_ap_miedo = corr_rectangular(
        base_corr,
        filas=bp_corr,
        columnas=ap_miedo,
        method="spearman"
    )
    
    plot_corr(
        corr_bp_ap_miedo,
        title=f"{year} - BP principales x AP miedo conductual",
        labels_dict=all_labels,
        figsize=(14, 6)
    )
    
    # --------------------------------------------------------
    # 3) BP nominales convertidas en dummies x AP percepción
    #    Lugar y daño principal.
    # --------------------------------------------------------
    corr_dummies_ap_percepcion = corr_rectangular(
        base_corr,
        filas=bp_dummies,
        columnas=ap_percepcion,
        method="spearman"
    )
    
    plot_corr(
        corr_dummies_ap_percepcion,
        title=f"{year} - Lugar/daño principal x AP percepción de inseguridad",
        labels_dict=all_labels,
        figsize=(12, 7)
    )
    
    # --------------------------------------------------------
    # 4) BP nominales convertidas en dummies x AP miedo conductual
    # --------------------------------------------------------
    corr_dummies_ap_miedo = corr_rectangular(
        base_corr,
        filas=bp_dummies,
        columnas=ap_miedo,
        method="spearman"
    )
    
    plot_corr(
        corr_dummies_ap_miedo,
        title=f"{year} - Lugar/daño principal x AP miedo conductual",
        labels_dict=all_labels,
        figsize=(15, 7)
    )
    
    # --------------------------------------------------------
    # 5) Matriz grande: todas las BP x todas las AP
    #    Sin anotaciones para que no quede ilegible.
    # --------------------------------------------------------
    corr_todas = corr_rectangular(
        base_corr,
        filas=bp_corr + bp_dummies,
        columnas=ap_percepcion + ap_miedo,
        method="spearman"
    )
    
    plot_corr(
        corr_todas,
        title=f"{year} - Todas las BP x todas las AP",
        labels_dict=all_labels,
        figsize=(18, 9),
        annot=False
    )
    
    # --------------------------------------------------------
    # 6) Resumen de correlaciones más fuertes
    # --------------------------------------------------------
    print("\nTop correlaciones absolutas BP x AP:")
    display(top_correlaciones(corr_todas, n=20))
```
