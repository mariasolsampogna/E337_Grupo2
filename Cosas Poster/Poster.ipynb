```python
# Para ver una matriz de correlación no nos sirve mucho hacerlo de manera directa porque son categorías binarias practicamente la mayoría de las varaibles que tenemos.
# Así que vamos a tener que hacer una transformación de algun tipo para que sean comparables.
```

```python
# Paso 1: Crear dummies solo para las variables seleccionadas
dummies = []

cols_interest = [
    "AP4_3_1",   # colonia
    "AP4_3_2",   # municipio
    "AP4_3_3",   # estado
    "AP4_4_03",  # calle
    "AP4_4_09",  # transporte público
    "AP4_4_A"    # caminar solo/a de noche
]

for col in cols_interest:
    df_col = df_analisis[col].astype("string")
    dummies.append(pd.get_dummies(df_col, prefix=col))

dummies = pd.concat(dummies, axis=1)
```

```python
# Paso 2: Renombrar con etiquetas legibles
rename_dict = {
    "AP4_3_1_1": "Colonia: Seguro",
    "AP4_3_1_2": "Colonia: Inseguro",

    "AP4_3_2_1": "Municipio: Seguro",
    "AP4_3_2_2": "Municipio: Inseguro",

    "AP4_3_3_1": "Estado: Seguro",
    "AP4_3_3_2": "Estado: Inseguro",

    "AP4_4_03_1": "Calle: Seguro",
    "AP4_4_03_2": "Calle: Inseguro",

    "AP4_4_09_1": "Transporte: Seguro",
    "AP4_4_09_2": "Transporte: Inseguro",

    "AP4_4_A_1": "Noche: Muy seguro",
    "AP4_4_A_2": "Noche: Seguro",
    "AP4_4_A_3": "Noche: Inseguro",
    "AP4_4_A_4": "Noche: Muy inseguro"
}

dummies = dummies.rename(columns=rename_dict)
```

```python
# Paso 3: Graficar
plt.figure(figsize=(14, 10))
sns.heatmap(dummies.corr(), vmin=-1, vmax=1, cmap="coolwarm")
plt.xticks(rotation=45, ha="right", fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.title("Matriz de correlación - Percepción de inseguridad")
plt.tight_layout()
plt.show()
```

```python
# =========================
# 1) BLOQUE DE MIEDO CONDUCTUAL (AP4_10)
# =========================
# Estas son las variables del bloque que querés reducir
cols_miedo = [f'AP4_10_{str(i).zfill(2)}' for i in range(1, 14)]

# Revisamos que existan en el dataframe
print("Variables encontradas en dfanalisis:")
print([c for c in cols_miedo if c in df_analisis.columns])

# Nos quedamos solo con las que realmente existan
cols_miedo = [c for c in cols_miedo if c in df_analisis.columns]
```

```python
# Reemplazamos códigos no válidos:
# 1 = Sí
# 2 = No
# 3 = No aplica
# 9 = No sabe / no responde
#
# Para que el índice tenga sentido:
# Sí -> 1
# No -> 0
# No aplica / NS-NR -> NaN
df_analisis[cols_miedo] = (
    df_analisis[cols_miedo]
    .replace({1: 1, 2: 0, 3: np.nan, 9: np.nan})
)

# Creamos índice de miedo conductual:
# promedio de respuestas válidas por persona
df_analisis["indice_miedo"] = df_analisis[cols_miedo].mean(axis=1)

print(df_analisis["indice_miedo"].describe())
```

```python
# =========================
# 2) PERCEPCIÓN DE INSEGURIDAD
# =========================
# Lugares donde se siente seguro/inseguro
cols_inseguridad = [f'AP4_4_{str(i).zfill(2)}' for i in range(1, 13)]

print("Variables AP4_4 encontradas:")
print([c for c in cols_inseguridad if c in df_analisis.columns])

cols_inseguridad = [c for c in cols_inseguridad if c in df_analisis.columns]

# Recodificación:
# 1 = Seguro(a)
# 2 = Inseguro(a)
# 3 = No aplica
# 9 = No sabe / no responde
#
# Queremos que mayor valor = más inseguridad
# Seguro -> 0
# Inseguro -> 1
# No aplica / NS-NR -> NaN
df_analisis[cols_inseguridad] = (
    df_analisis[cols_inseguridad]
    .replace({1: 0, 2: 1, 3: np.nan, 9: np.nan})
)

# Índice promedio de inseguridad en distintos espacios
df_analisis["indice_inseguridad_espacios"] = df_analisis[cols_inseguridad].mean(axis=1)

# Variable puntual: caminar sola/o de noche
# AP4_4_A:
# 1 Muy seguro(a), 2 Seguro(a), 3 Inseguro(a), 4 Muy inseguro(a), 5 No aplica, 9 NS/NR
if "AP4_4_A" in df_analisis.columns:
    df_analisis["AP4_4_A_rec"] = df_analisis["AP4_4_A"].replace({
        1: 0,   # muy seguro
        2: 1,   # seguro
        3: 2,   # inseguro
        4: 3,   # muy inseguro
        5: np.nan,
        9: np.nan
    })

print(df_analisis["indice_inseguridad_espacios"].describe())

if "AP4_4_A_rec" in df_analisis.columns:
    print(df_analisis["AP4_4_A_rec"].describe())
```

```python
# =========================
# 3) CONFIANZA EN INSTITUCIONES
# =========================
# AP5_4_01 a AP5_4_11
cols_confianza = [f'AP5_4_{str(i).zfill(2)}' for i in range(1, 12)]

print("Variables AP5_4 encontradas:")
print([c for c in cols_confianza if c in df_analisis.columns])

cols_confianza = [c for c in cols_confianza if c in df_analisis.columns]

# Supuesto de codificación:
# 1 = mucha confianza
# 2 = algo de confianza
# 3 = algo de desconfianza
# 4 = mucha desconfianza
# otros códigos raros/blancos -> NaN
#
# Si querés que mayor valor = MENOR confianza, lo dejamos así.
df_analisis[cols_confianza] = df_analisis[cols_confianza].replace({
    8: np.nan, 9: np.nan, "b": np.nan, " ": np.nan, "": np.nan
})

df_analisis["indice_desconfianza"] = df_analisis[cols_confianza].mean(axis=1)

# Si preferís que mayor valor = MAYOR confianza, invertimos la escala:
df_analisis["indice_confianza"] = df_analisis["indice_desconfianza"].replace({
    1: 4,
    2: 3,
    3: 2,
    4: 1
})

print(df_analisis["indice_desconfianza"].describe())
print(df_analisis["indice_confianza"].describe())
```

```python
# =========================
# 4) MATRIZ DE CORRELACIONES REDUCIDA
# =========================
vars_corr = [
    "indice_confianza",
    "indice_desconfianza",
    "indice_inseguridad_espacios",
    "indice_miedo"
]

# Agregamos AP4_4_A si está
if "AP4_4_A_rec" in df_analisis.columns:
    vars_corr.append("AP4_4_A_rec")

# Nos quedamos solo con las que existan de verdad
vars_corr = [c for c in vars_corr if c in df_analisis.columns]

corr_df = df_analisis[vars_corr].corr(numeric_only=True)

print(corr_df.round(3))
```

```python
# =========================
# 5) HEATMAP LIMPIO
# =========================
plt.figure(figsize=(8, 6))
sns.heatmap(corr_df, annot=False, cmap="coolwarm", vmin=-1, vmax=1, square=True)
plt.title("Matriz de correlaciones reducida")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Construimos la variable victima usando AP7_1 (victimización en el hogar o personal)
# AP7_1: 1 = fue víctima, 2 = no fue víctima
df_TPer_Vic2['AP7_1'] = df_TPer_Vic2['AP7_1'].replace([9, 'b'], np.nan)
df_TPer_Vic2['victima'] = df_TPer_Vic2['AP7_1'].map({1: 1, 2: 0})

print(df_TPer_Vic2['victima'].value_counts(dropna=False))
```

```python
# Limpieza de variables de hábitos en la base original ANTES del merge
df_TPer_Vic2[cols_habitos] = df_TPer_Vic2[cols_habitos].replace([9, 'b'], np.nan)
for col in cols_habitos:
    df_TPer_Vic2[col] = df_TPer_Vic2[col].map({1: 1, 2: 0})
```

```python
# Merge con df_analisis 
df_analisis_p3 = df_analisis.merge(
    df_TPer_Vic2[['UPM', 'VIV_SEL', 'HOGAR', 'R_SEL', 'victima'] + cols_habitos],
    on=['UPM', 'VIV_SEL', 'HOGAR', 'R_SEL'],
    how='left'
)

print(f"Filas df_analisis:    {len(df_analisis)}")
print(f"Filas df_analisis_p3: {len(df_analisis_p3)}")
```

```python
# fillna(0): quien no respondió estas preguntas no cambió ese hábito
for col in cols_habitos:
    df_analisis_p3[col] = df_analisis_p3[col].fillna(0)
```

```python
# Proporción general de cambio de hábitos en toda la muestra
proporcion_general = df_analisis_p3[cols_habitos].mean() * 100
proporcion_general = proporcion_general.rename(labels_habitos).sort_values(ascending=True)

plt.figure(figsize=(8, 4))
sns.barplot(x=proporcion_general.values, y=proporcion_general.index, color='steelblue')
plt.xlabel("% que modificó el hábito")
plt.ylabel("")
plt.title("Proporción general de cambio de hábitos por temor al delito")
plt.tight_layout()
plt.show()
```

```python
# Estadística descriptiva: % que modificó cada hábito por condición de víctima
tabla_desc = df_analisis_p3.groupby('victima')[cols_habitos].mean() * 100
tabla_desc = tabla_desc.rename(columns=labels_habitos)
tabla_desc.index = tabla_desc.index.map({0: 'No víctima', 1: 'Víctima'})
tabla_desc
```

```python
# Distribución del índice para víctimas vs no víctimas
fig, ax = plt.subplots(figsize=(8, 5))

for valor, color, label in [(0, 'steelblue', 'No víctima'), (1, 'tomato', 'Víctima')]:
    subset = df_analisis_p3[df_analisis_p3['victima'] == valor]['indice_habitos'].value_counts().sort_index()
    ax.bar(subset.index + (0.2 if valor == 1 else -0.2), subset.values, 
           width=0.35, color=color, label=label)

ax.set_xticks([0, 1, 2, 3])
ax.set_xlabel("Número de hábitos modificados (0 a 3)")
ax.set_ylabel("Frecuencia")
ax.set_title("Cantidad de hábitos modificados por miedo al delito\nVíctimas vs. No víctimas")
ax.legend(title='Condición')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras horizontales: % que modificó cada hábito, víctimas vs no víctimas
tabla_plot = tabla_desc.T.reset_index()
tabla_plot.columns = ['Hábito', 'No víctima', 'Víctima']
tabla_plot = tabla_plot.sort_values('Víctima', ascending=True)

fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(len(tabla_plot))
width = 0.35

ax.barh(x - width/2, tabla_plot['No víctima'], width, label='No víctima', color='steelblue')
ax.barh(x + width/2, tabla_plot['Víctima'],    width, label='Víctima',    color='tomato')

ax.set_yticks(x)
ax.set_yticklabels(tabla_plot['Hábito'], fontsize=10)
ax.set_xlabel("% que modificó el hábito")
ax.set_title("Cambio de hábitos por temor al delito\nVíctimas vs. No víctimas")
ax.legend()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, _: f'{val:.0f}%'))
plt.tight_layout()
plt.show()
```
