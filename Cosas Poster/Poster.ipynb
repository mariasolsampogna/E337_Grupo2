```python
# =========================================================
# 1. DEFINIR BLOQUES DE VARIABLES
# =========================================================
cols_total = cols_conf + cols_corr + cols_efec

# =========================================================
# 2. DICCIONARIO DE LABELS ABREVIADOS
# =========================================================
# Reemplazá los nombres de las autoridades por los correctos
# según tu base / cuestionario

authority_labels = {
    "01": "Autoridad 1",
    "02": "Autoridad 2",
    "03": "Autoridad 3",
    "04": "Autoridad 4",
    "05": "Autoridad 5",
    "06": "Autoridad 6",
    "07": "Autoridad 7",
    "08": "Autoridad 8",
    "09": "Autoridad 9",
    "10": "Autoridad 10",
    "11": "Autoridad 11"
}

# Armamos automáticamente labels abreviados para cada bloque
labels = {}

# Confianza institucional
for key, val in authority_labels.items():
    labels[f"AP5_4_{key}"] = f"Conf: {val}"

# Corrupción percibida
for key, val in authority_labels.items():
    labels[f"AP5_5_{key}"] = f"Corr: {val}"

# Efectividad percibida
for key, val in authority_labels.items():
    labels[f"AP5_6_{key}"] = f"Efec: {val}"


# =========================================================
# 3. FUNCIÓN PARA GRAFICAR HEATMAP DE MISSING POR BLOQUE
# =========================================================
os.makedirs("graficos", exist_ok=True)

def plot_missing_block(df, cols, labels_dict, title, filename, figsize=(12, 2.8), annot=True, sort=True):
    
    missing_pct = df[cols].isnull().mean() * 100
    
    if sort:
        missing_pct = missing_pct.sort_values(ascending=False)
    
    missing_df = pd.DataFrame(missing_pct).T
    missing_df.columns = [labels_dict.get(col, col) for col in missing_df.columns]
    
    plt.figure(figsize=figsize)
    
    ax = sns.heatmap(
        missing_df,
        annot=annot,
        fmt=".1f",
        cmap="Greens",
        vmin=0,
        vmax=100,
        linewidths=0.5,
        cbar_kws={"label": "% Missing"}
    )
    
    ax.set_yticklabels(["% Missing"], rotation=0)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=9)
    
    plt.title(title)
    plt.tight_layout()
    
    plt.savefig(f"graficos/{filename}.png", dpi=300, bbox_inches="tight")
    plt.show()
# =========================================================
# 4. GRAFICAR CADA BLOQUE POR SEPARADO
# =========================================================

plot_missing_block(
    df=df_analisis,
    cols=cols_conf,
    labels_dict=labels,
    title="Porcentaje de valores faltantes - Confianza institucional",
    filename="missing_confianza_2025",
    figsize=(12, 2.8),
    annot=True
)

plot_missing_block(
    df=df_analisis,
    cols=cols_corr,
    labels_dict=labels,
    title="Porcentaje de valores faltantes - Corrupción percibida",
    filename="missing_corrupcion_2025",
    figsize=(12, 2.8),
    annot=True
)

plot_missing_block(
    df=df_analisis,
    cols=cols_efec,
    labels_dict=labels,
    title="Porcentaje de valores faltantes - Efectividad percibida",
    filename="missing_efectividad_2025",
    figsize=(12, 2.8),
    annot=True
)
```

```python
# Paso 1: Definir y estandarizar las variables por dimensión
 
# Dimensión A: Percepción de inseguridad geográfica (AP4_3) 
# 1 = seguro, 2 = inseguro → recodificamos a 0/1
cols_geo = ["AP4_3_1", "AP4_3_2", "AP4_3_3"]
 
df_analisis[cols_geo] = (
    df_analisis[cols_geo]
    .replace({9: np.nan})
    .apply(pd.to_numeric, errors="coerce")
)
# Convertimos a binaria 0=seguro, 1=inseguro
for col in cols_geo:
    df_analisis[col] = df_analisis[col].map({1: 0, 2: 1})
 
# Índice: promedio de las 3 variables geográficas (0=nada inseguro, 1=todo inseguro)
df_analisis["idx_inseg_geo"] = df_analisis[cols_geo].mean(axis=1)
```

```python
# Dimensión B: Percepción de inseguridad en espacios (AP4_4_01 a AP4_4_12 + AP4_4_A) 
# AP4_4_01..12: 1=seguro, 2=inseguro (ya limpiado antes con replace([3,9], nan))
# Recodificamos a 0/1
cols_espacios = [f"AP4_4_{str(i).zfill(2)}" for i in range(1, 13)]
cols_espacios = [c for c in cols_espacios if c in df_analisis.columns]
 
for col in cols_espacios:
    df_analisis[col] = (
        pd.to_numeric(df_analisis[col], errors="coerce")
        .map({1: 0, 2: 1})
    )
```

```python
# =========================
# 2) PERCEPCIÓN DE INSEGURIDAD
# =========================
# Lugares donde se siente seguro/inseguro
cols_inseguridad = [f'AP4_4_{str(i).zfill(2)}' for i in range(1, 13)]

print("Variables AP4_4 encontradas:")
print([c for c in cols_inseguridad if c in df_analisis.columns])

cols_inseguridad = [c for c in cols_inseguridad if c in df_analisis.columns]

# Recodificación:
# 1 = Seguro(a)
# 2 = Inseguro(a)
# 3 = No aplica
# 9 = No sabe / no responde
#
# Queremos que mayor valor = más inseguridad
# Seguro -> 0
# Inseguro -> 1
# No aplica / NS-NR -> NaN
df_analisis[cols_inseguridad] = (
    df_analisis[cols_inseguridad]
    .replace({1: 0, 2: 1, 3: np.nan, 9: np.nan})
)

# Índice promedio de inseguridad en distintos espacios
df_analisis["indice_inseguridad_espacios"] = df_analisis[cols_inseguridad].mean(axis=1)

# Variable puntual: caminar sola/o de noche
# AP4_4_A:
# 1 Muy seguro(a), 2 Seguro(a), 3 Inseguro(a), 4 Muy inseguro(a), 5 No aplica, 9 NS/NR
if "AP4_4_A" in df_analisis.columns:
    df_analisis["AP4_4_A_rec"] = df_analisis["AP4_4_A"].replace({
        1: 0,   # muy seguro
        2: 1,   # seguro
        3: 2,   # inseguro
        4: 3,   # muy inseguro
        5: np.nan,
        9: np.nan
    })

print(df_analisis["indice_inseguridad_espacios"].describe())

if "AP4_4_A_rec" in df_analisis.columns:
    print(df_analisis["AP4_4_A_rec"].describe())

# AP4_4_A: 1=muy seguro, 2=seguro, 3=inseguro, 4=muy inseguro
# Normalizamos a [0,1] dividiendo por 3
if "AP4_4_A" in df_analisis.columns:
    df_analisis["AP4_4_A_norm"] = (
        pd.to_numeric(df_analisis["AP4_4_A"], errors="coerce")
        .replace({5: np.nan, 9: np.nan})
        .sub(1)          # rango 0-3
        .div(3)          # escala 0-1
    )
    cols_espacios_full = cols_espacios + ["AP4_4_A_norm"]
else:
    cols_espacios_full = cols_espacios
 
df_analisis["idx_inseg_espacios"] = df_analisis[cols_espacios_full].mean(axis=1)
```

```python
# Dimensión C: Confianza en instituciones (AP5_4)
# 1=mucha confianza → 4=mucha desconfianza
# Normalizamos a [0,1]: (valor-1)/3 → 0=mucha confianza, 1=mucha desconfianza
cols_conf_all = [c for c in df_analisis.columns if c.startswith("AP5_4_")]
 
for col in cols_conf_all:
    df_analisis[col] = (
        pd.to_numeric(df_analisis[col], errors="coerce")
        .replace({8: np.nan, 9: np.nan})
        .sub(1)
        .div(3)
    )
 
df_analisis["idx_desconfianza"] = df_analisis[cols_conf_all].mean(axis=1)
 
 
# --- Dimensión D: Percepción de corrupción (AP5_5) ---
# 1=Sí es corrupta, 2=No → recodificamos a 1=corrupta, 0=no corrupta
cols_corr_all = [c for c in df_analisis.columns if c.startswith("AP5_5_")]
 
for col in cols_corr_all:
    df_analisis[col] = (
        pd.to_numeric(df_analisis[col], errors="coerce")
        .replace({9: np.nan})
        .map({1: 1, 2: 0})
    )
 
df_analisis["idx_corrupcion"] = df_analisis[cols_corr_all].mean(axis=1)
```

```python
# Dimensión E: Percepción de inefectividad (AP5_6)
# 1=muy efectivo → 4=nada efectivo
# Normalizamos a [0,1]: (valor-1)/3 → 0=muy efectivo, 1=nada efectivo
cols_efec_all = [c for c in df_analisis.columns if c.startswith("AP5_6_")]
 
for col in cols_efec_all:
    df_analisis[col] = (
        pd.to_numeric(df_analisis[col], errors="coerce")
        .replace({9: np.nan})
        .sub(1)
        .div(3)
    )
 
df_analisis["idx_inefectividad"] = df_analisis[cols_efec_all].mean(axis=1)
 
 
# --- Dimensión F: Miedo conductual (AP4_10) ---
# 1=Sí dejó de hacer algo, 2=No, 3=No aplica, 9=NS-NR
cols_miedo = [c for c in df_analisis.columns
              if c.startswith("AP4_10_") and c in df_analisis.columns]
 
for col in cols_miedo:
    df_analisis[col] = (
        pd.to_numeric(df_analisis[col], errors="coerce")
        .replace({3: np.nan, 9: np.nan})
        .map({1: 1, 2: 0})
    )
 
df_analisis["idx_miedo_conductual"] = df_analisis[cols_miedo].mean(axis=1)
```

```python
# PASO 2: Verificar los índices

indices = [
    "idx_inseg_geo",
    "idx_inseg_espacios",
    "idx_desconfianza",
    "idx_corrupcion",
    "idx_inefectividad",
    "idx_miedo_conductual",
]
 
print("Estadísticas de los índices (escala 0-1):")
print(df_analisis[indices].describe().round(3))
print()
print("Cantidad de no-nulos por índice:")
print(df_analisis[indices].notna().sum())
```

```python
# Verificación de los índices construidos
 
indices_2024 = [
    "idx_inseg_geo",
    "idx_inseg_espacios",
    "idx_desconfianza",
    "idx_corrupcion",
    "idx_inefectividad",
    "idx_miedo_conductual",
]
 
print("Estadísticas de los índices 2024 (escala 0-1):")
print(df_analisis_2024[indices_2024].describe().round(3))
print()
print("Cantidad de no-nulos por índice:")
print(df_analisis_2024[indices_2024].notna().sum())
```
