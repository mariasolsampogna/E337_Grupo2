```python
# Para poder ver esto, vamos a neceditar distintas bases, siendo estas df_TSDem y df_TPer_Vic1
# Antes de eso, estaria bueno guardar todo dentro de regiones en vez de tener todos los estados por separado.
```

```python
df_TSDem['NOM_ENT']
```

```python
# Estandarizamos a los nombres por las dudas
df_TSDem["NOM_ENT"] = df_TSDem["NOM_ENT"].str.strip().str.upper()
df_TPer_Vic1["NOM_ENT"] = df_TPer_Vic1["NOM_ENT"].str.strip().str.upper()

# Creamos un diccionario para poder identificar mejor cada estado de México
regiones_mexico = {
    "Noroeste": [
        "BAJA CALIFORNIA", "BAJA CALIFORNIA SUR", "CHIHUAHUA",
        "DURANGO", "SINALOA", "SONORA"
    ],
    "Noreste": [
        "COAHUILA DE ZARAGOZA", "NUEVO LEON", "TAMAULIPAS"
    ],
    "Occidente": [
        "COLIMA", "GUANAJUATO", "JALISCO",
        "MICHOACAN DE OCAMPO", "NAYARIT"
    ],
    "Centronorte": [
        "AGUASCALIENTES", "QUERETARO",
        "SAN LUIS POTOSI", "ZACATECAS"
    ],
    "Centrosur": [
        "CIUDAD DE MEXICO", "MEXICO", "MORELOS"
    ],
    "Oriente": [
        "HIDALGO", "PUEBLA", "TLAXCALA",
        "VERACRUZ DE IGNACIO DE LA LLAVE"
    ],
    "Suroeste": [
        "CHIAPAS", "GUERRERO", "OAXACA"
    ],
    "Sureste": [
        "CAMPECHE", "QUINTANA ROO",
        "TABASCO", "YUCATAN"
    ]
}
```

```python
# Creamos una función que nos facilite agrupar a los estados
def asignar_region(estado):
    for region, estados in regiones_mexico.items():
        if estado in estados:
            return region
    return "No clasificado"

# Aplicamos la función construida
df_TSDem["region"] = df_TSDem["NOM_ENT"].apply(asignar_region)
df_TPer_Vic1["region"] = df_TPer_Vic1["NOM_ENT"].apply(asignar_region)
```

```python
print(df_TSDem["region"].value_counts())
```

```python
df_TSDem[df_TSDem["region"] == "No clasificado"]["NOM_ENT"].drop_duplicates()
```

```python
# Ahora hacemos merge para todo esto
df_analisis = df_TPer_Vic1.merge(
    df_TSDem[["UPM", "VIV_SEL", "HOGAR", "N_REN", "SEXO", "EDAD", "region"]],
    left_on=["UPM", "VIV_SEL", "HOGAR", "R_SEL"],
    right_on=["UPM", "VIV_SEL", "HOGAR", "N_REN"],
    how="inner",
    suffixes=("_vic", "_dem")   
)

# Eliminamos las columnas duplicadas del lado de victimización
# y renombramos las demográficas para que queden limpias
df_analisis = df_analisis.drop(columns=["SEXO_vic", "EDAD_vic", "region_vic"], errors="ignore")
df_analisis = df_analisis.rename(columns={
    "SEXO_dem": "SEXO",
    "EDAD_dem": "EDAD",
    "region_dem": "region"
})
```

```python
# Verificación
print(df_analisis[["SEXO", "EDAD", "region"]].head())
print(df_analisis["region"].value_counts())
```

```python
# Una vez con esto construido, podemos ahora si apreciar que variables nos pueden servir para esta pregunta. 
# Por un lado dentro de df_TPer_Vic1, vamos a necesitar a aquellas variables que nos indiquen la persepción general de manera geográfica, cosa que lo hace 
# las varaibles AP4_3_1 (colonia/localidad (1=seguro, 2=inseguro)), AP4_3_2 (municipio (1=seguro, 2=inseguro)), AP4_3_3 (estado (1=seguro, 2=inseguro))
```

```python
# Por otro lado, tenemos AP4_4 la sensación de seguridad en lugares específicos, siendo que 
# AP4_4_01 a AP4_4_12 → casa, trabajo, calle, escuela, mercado, centro comercial, banco, cajero, transporte público, auto, carretera, parque.
# AP4_4_A → qué tan seguro se siente caminando solo de noche cerca de su vivienda (escala de 4 puntos: muy seguro → muy inseguro)
```

```python
# También AP4_7 como expectativa futura de seguridad, siendo que AP4_7_1, AP4_7_2, AP4_7_3 → si cree que la seguridad mejorará/empeorará en colonia, estado y país
```

```python
# Pero, tengamos cuidado con el como se calsificaron los NonValues, porque como nos paso con la EPH esto ocurre para algunas variables
```

```python
# En AP4_3, AP4_4: 9 → "No sabe / no responde", 3 → "No aplica" (en AP4_4, por ejemplo alguien que no tiene auto no responde AP4_4_10)
# En AP4_4_A: 5 → "No aplica", 9 → "No sabe / no responde"
# Para las variables geográficas como EDAD: 98 → edad no especificada (este sí es missing)
```

```python
# Variables de percepción por ámbito (AP4_3_1, AP4_3_2, AP4_3_3)
cols_ambito = ["AP4_3_1", "AP4_3_2", "AP4_3_3"]
df_analisis[cols_ambito] = df_analisis[cols_ambito].replace([9], np.nan)

# Variables de percepción por lugar (AP4_4_01 a AP4_4_12)
cols_lugares = [f"AP4_4_{str(i).zfill(2)}" for i in range(1, 13)]
df_analisis[cols_lugares] = df_analisis[cols_lugares].replace([3, 9], np.nan)

# AP4_4_A (caminar solo de noche)
df_analisis["AP4_4_A"] = df_analisis["AP4_4_A"].replace([5, 9], np.nan)
df_analisis["EDAD"] = df_analisis["EDAD"].replace([98], np.nan)
```

```python
# Para ver una matriz de correlación no nos sirve mucho hacerlo de manera directa porque son categorías binarias practicamente la mayoría de las varaibles que tenemos.
# Así que vamos a tener que hacer una transformación de algun tipo para que sean comparables.
```

```python
# Paso 1: Crear dummies para las variables categóricas
dummies = []

# SEXO
df_sexo = df_analisis["SEXO"].astype("string")
dummies.append(pd.get_dummies(df_sexo, prefix="SEXO"))

# REGION
df_region = df_analisis["region"].astype("string")
dummies.append(pd.get_dummies(df_region, prefix="REGION"))

# Definamos las variables de interes:

cols_interest = [
    "AP4_3_1", "AP4_3_2", "AP4_3_3",
    "AP4_4_01", "AP4_4_02", "AP4_4_03", "AP4_4_04",
    "AP4_4_05", "AP4_4_06", "AP4_4_07", "AP4_4_08",
    "AP4_4_09", "AP4_4_10", "AP4_4_11", "AP4_4_12",
    "AP4_4_A"
]

for col in cols_interest:
    df_col = df_analisis[col].astype("string")
    dummies.append(pd.get_dummies(df_col, prefix=col))

# EDAD va directa, es numérica
dummies.append(df_analisis[["EDAD"]])

dummies = pd.concat(dummies, axis=1)
```

```python
# Paso 2: Renombrar con etiquetas legibles
rename_dict = {
    # SEXO
    "SEXO_1": "Sexo: Hombre",
    "SEXO_2": "Sexo: Mujer",

    # REGION
    "REGION_Noroeste": "Región: Noroeste",
    "REGION_Noreste": "Región: Noreste",
    "REGION_Occidente": "Región: Occidente",
    "REGION_Centronorte": "Región: Centronorte",
    "REGION_Centrosur": "Región: Centrosur",
    "REGION_Oriente": "Región: Oriente",
    "REGION_Suroeste": "Región: Suroeste",
    "REGION_Sureste": "Región: Sureste",

    # Percepción por ámbito
    "AP4_3_1_1": "Colonia: Seguro",
    "AP4_3_1_2": "Colonia: Inseguro",
    "AP4_3_2_1": "Municipio: Seguro",
    "AP4_3_2_2": "Municipio: Inseguro",
    "AP4_3_3_1": "Estado: Seguro",
    "AP4_3_3_2": "Estado: Inseguro",

    # Percepción por lugar
    "AP4_4_01_1": "Casa: Seguro", "AP4_4_01_2": "Casa: Inseguro",
    "AP4_4_02_1": "Trabajo: Seguro", "AP4_4_02_2": "Trabajo: Inseguro",
    "AP4_4_03_1": "Calle: Seguro", "AP4_4_03_2": "Calle: Inseguro",
    "AP4_4_04_1": "Escuela: Seguro", "AP4_4_04_2": "Escuela: Inseguro",
    "AP4_4_05_1": "Mercado: Seguro", "AP4_4_05_2": "Mercado: Inseguro",
    "AP4_4_06_1": "C.Comercial: Seguro", "AP4_4_06_2": "C.Comercial: Inseguro",
    "AP4_4_07_1": "Banco: Seguro", "AP4_4_07_2": "Banco: Inseguro",
    "AP4_4_08_1": "Cajero: Seguro", "AP4_4_08_2": "Cajero: Inseguro",
    "AP4_4_09_1": "Transporte: Seguro", "AP4_4_09_2": "Transporte: Inseguro",
    "AP4_4_10_1": "Auto: Seguro", "AP4_4_10_2": "Auto: Inseguro",
    "AP4_4_11_1": "Carretera: Seguro", "AP4_4_11_2": "Carretera: Inseguro",
    "AP4_4_12_1": "Parque: Seguro", "AP4_4_12_2": "Parque: Inseguro",

    # Caminar de noche
    "AP4_4_A_1": "Noche: Muy seguro",
    "AP4_4_A_2": "Noche: Seguro",
    "AP4_4_A_3": "Noche: Inseguro",
    "AP4_4_A_4": "Noche: Muy inseguro",

    # EDAD
    "EDAD": "Edad"
}

dummies = dummies.rename(columns=rename_dict)
```

```python
# Paso 3: Abreviar
dummies.columns = dummies.columns.str.replace("Región: ", "Reg: ", regex=False)
dummies.columns = dummies.columns.str.replace("Sexo: ", "Sex: ", regex=False)

# Paso 4: Graficar
plt.figure(figsize=(24, 20))
sns.heatmap(dummies.corr(), vmin=-1, vmax=1, cmap="coolwarm")
plt.xticks(rotation=45, ha="right", fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.title("Matriz de correlación - Percepción de inseguridad")
plt.tight_layout()
plt.show()
```

```python
# Dividimos variables
cols_1 = ["AP4_3_1", "AP4_3_2", "AP4_3_3",
          "AP4_4_01", "AP4_4_02", "AP4_4_03"]

cols_2 = ["AP4_4_04", "AP4_4_05", "AP4_4_06",
          "AP4_4_07", "AP4_4_08", "AP4_4_09",
          "AP4_4_10", "AP4_4_11", "AP4_4_12", "AP4_4_A"]

# Correlaciones
corr1 = df_analisis[cols_1].corr(method="spearman")
corr2 = df_analisis[cols_2].corr(method="spearman")

# Crear figura con dos paneles
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Matriz 1
sns.heatmap(
    corr1,
    annot=False,
    cmap="Oranges",
    vmin=-1, vmax=1,
    ax=axes[0]
)
axes[0].set_title("Correlaciones - Bloque 1")
axes[0].tick_params(axis='x', rotation=45)

# Matriz 2
sns.heatmap(
    corr2,
    annot=False,
    cmap="Greens",
    vmin=-1, vmax=1,
    ax=axes[1]
)
axes[1].set_title("Correlaciones - Bloque 2")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
